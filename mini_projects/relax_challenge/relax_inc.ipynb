```python
# Create a rolling 7day window
def rolling(dataframe, freq):
    return dataframe.rolling(freq, on='time_stamp')['visited'].count()

user_engagement['visits_in7days'] = user_engagement.groupby('user_id', as_index=False, group_keys=False).apply(rolling, freq='7D')

user_engagement.head()
```

```python
# adopted users
adopted = user_engagement.loc[(user_engagement['visits_in7days'] >= 3)]
adopted = adopted.drop_duplicates('user_id', keep = 'first')
print("Users who adopted: {}".format(len(adopted)))
adopted.head()
```

```python
# Add a column 'adopted' to dataframe users
adopted_users = list(adopted['user_id'])
adopted_users

users['adopted'] = users['object_id'].isin(adopted_users)
users.head()
```

```python
# Derive new features from columns creation_time and last_session_create_time
from datetime import datetime, timedelta

max_last_session_time = max(users['last_session_creation_time'])

diff_last_session_to_creation_time = (users['last_session_creation_time'] - pd.to_datetime(users['creation_time'])).dt.days

diff_max_last_session_to_session_time = (max_last_session_time - users['last_session_creation_time']).dt.days
```

```python
# Add the new features to dataframe users
users['diff_lastsession_to_creation'] = diff_last_session_to_creation_time
users['diff_maxlastsession_to_lastsession'] = diff_max_last_session_to_session_time

users.head()
```

```python
# Remove redundant columns 
users_mod = users.drop(columns=['name','creation_time','last_session_creation_time'])
users_mod.head()
```

```python
# Extract email domain name from email address
users_mod['email'] = [x.split('@')[1] for x in users.email]
users_mod.head()
```

```python
# Do one hot encoding to convert categorical varibles into numerical equivalents
users_mod_encoded = pd.get_dummies(users_mod)
users_mod_encoded.head()
```

```python
# Find important features using Logistic regression

# Split dataset into training set and test set
from sklearn.model_selection import train_test_split

x = users_mod_encoded[users_mod_encoded.columns]
y = users_mod_encoded['adopted']
x.drop(columns=['adopted'], inplace=True)
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Assess the relative importance of features by using logistic regression and looking at coefficients of 
# features/independent variables
from sklearn.linear_model import LogisticRegression

classifier_logistic = LogisticRegression(solver='liblinear')
classifier_logistic.fit(x_train, y_train)

df_coefs = pd.DataFrame(np.round(classifier_logistic.coef_[0], 3))
df_coefs.index = x_train.columns
df_coefs_sorted = df_coefs.sort_values(by=0)
df_coefs_sorted
```
